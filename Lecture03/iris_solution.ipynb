# Iris

The Iris dataset was used in R.A. Fisher's classic 1936 paper, The Use of Multiple Measurements in Taxonomic Problems, and can also be found on the UCI Machine Learning Repository.

It includes three iris species with 50 samples each as well as some properties about each flower. One flower species is linearly separable from the other two, but the other two are not linearly separable from each other.

The columns in this dataset are:

- Id
- SepalLength, Cm
- SepalWidth, Cm
- PetalLength, Cm
- PetalWidth, Cm
- Species

## Import data

```python
iris = pd.read_csv('./data/iris.csv', index_col='Id')
```

```python
iris.head()
```

## Boxplot

```python
iris.boxplot(grid=False, figsize=[7, 7]);
```

Observationer:
- Max-, min- och medelvärde
- 25th och 75th percentile
- Man ser hur datan är fördelad, också i förhållande till medelvärdet
- Finns det outliers kan det bli synligt här. Denna datan ser dock fin ut


## Histogram

```python
iris.hist(grid=False, figsize=[7, 7]);
```

Observationer:
- Histogram visar distributionen  (fördelningen) till varje kolumn
- Petal length och width har inte symmetrisk distribution. Flera observationer med små storleker som förklarar box ploten för petal length
- Sepal width ser normalfördelad ut


## Scatter plot

```python
pd.plotting.scatter_matrix(iris, figsize=[10, 10]);
```

Observationer:
- Två och två kolumner är plottet mot varandra. Övre diagonal är därmed spegelbild av nedre diagonal 
- Man ser tydiga kluster i petal width vs petal length. Dessa två variabler förklarar antagligen datan väl
- Setosa skiljer sig tydligt från de andra två arterna
